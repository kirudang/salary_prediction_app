Multiple steps will be done in this part to ensure data quality for modeling.

```python
# create a grid of subplots with one row and two columns
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 5))

# create the histogram of salary using Seaborn
sns.histplot(ax=ax1, data=df, x='Salary', kde=True, bins=20)

# add axis labels and a title to the histogram plot
ax1.set_xlabel('Salary')
ax1.set_ylabel('Frequency')
ax1.set_title('Histogram of Salary')

# create the boxplot of salary using Seaborn
sns.boxplot(ax=ax2, data=df, y='Salary')

# add axis labels and a title to the boxplot plot
ax2.set_xlabel('Salary')
ax2.set_ylabel('Salary')
ax2.set_title('Boxplot of Salary')

# adjust the spacing between subplots
plt.tight_layout()

# show the plot
plt.show()
```

```python
df["Salary"].describe()
```

As we can see, Salary is highly skewed with min of **1** USD per year while the max is up to **50,000,000**.
Keep these data will make our machine learning does not generalize well, hence produces poor predictions.

```python
q25_salary = df['Salary'].quantile(0.025)
q975_salary = df['Salary'].quantile(0.975)
print("2.5 percentile of Salary is " + "${:,.02f}".format(q25_salary))
print("97.5 percentile of Salary is " + "${:,.02f}".format(q975_salary))
```

```python
# Filter data, only keep Salary from 1% to 98% quantile
df = df[(df["Salary"]>= q25_salary) & (df["Salary"]<= q975_salary)]
```

```python
# Check dimension of data
df.shape
```

As of this step, our data remain **25,594** observations, which are enough to train.

```python
# Drop fully duplicated rows
df.drop_duplicates(keep='first', inplace=True)
```

#### - Feature transformation
Each function will:

    - Convert data into suitable format
    - Bundle data

- **COUNTRY**

```python
# Top 15 countries with highest frequency
top_country = df["Country"].value_counts(normalize = True).head(15)
top_country
```

```python
sum(top_country)
```

As we have top 15 countries accounting for almost 70% of data, we will bundle this categorical variables, in which countries out of these 15 mentioned ones will be listed as "Others".

```python
# Top countries
top_country_names = top_country.index.tolist()
top_country_names
```

```python
# Define the function
def clean_country(x):
    if x in top_country_names:
        return x
    else:
        return "Others"

# Apply the function to the 'Country' column in the DataFrame
df['Country'] = df['Country'].apply(clean_country)
```

```python
# Plot
country_mean_salary = df.groupby('Country')['Salary'].mean().sort_values(ascending=False)

plt.bar(country_mean_salary.index, country_mean_salary.values)
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Average Salary')
plt.title('Average Salary by Country')
plt.show()
```

- **EDUCATION**

```python
# Elements of Education
df["Education"].value_counts(normalize = True).head(15)
```

As we have many levels of Education, the author bundles them into smaller set.

```python
# Create a function to clean data
def clean_education(x):
    if 'Bachelor’s degree' in x:
        return 'Bachelor’s degree'
    if 'Master’s degree' in x:
        return 'Master’s degree'
    if 'Professional degree' in x or 'Other doctoral' in x:
        return 'Higher education(PhD, Doctoral degrees,...)'
    return 'Less than a Bachelors'

df['Education'] = df['Education'].apply(clean_education)
```

```python
# Elements of Education
df["Education"].value_counts(normalize = True).head(15)
```

```python
# Plot create the boxplot of salary by education level using Seaborn
sns.boxplot(x='Education', y='Salary', data=df)
plt.xticks(rotation=90)
# add axis labels and a title to the plot
plt.xlabel('Education Level')
plt.ylabel('Salary')
plt.title('Salary by Education Level')

# show the plot
plt.show()
```

- **CODING EXPERIENCE**

```python
df['Year_of_Code'].unique()
```

Since we have our coding experience data in object format, we need to convert them into number.

```python
# Create function
def clean_coding_experience(x):
    if x ==  'More than 50 years':
        return 51
    if x == 'Less than 1 year':
        return 0.5
    return float(x)

df['Year_of_Code'] = df['Year_of_Code'].apply(clean_coding_experience)
```

```python
# create the histogram of salary using Seaborn
sns.histplot(data=df, x='Year_of_Code', kde=True)

# add axis labels and a title to the histogram plot
plt.xlabel('Coding Experience')
plt.ylabel('Frequency')
plt.title('Histogram of Coding Experience')

# show the plot
plt.show()
```

- **EMPLOYMENT TYPE**

```python
df['Employment'].value_counts(normalize = True)
```

From the above classification, the author will combine data into 4 groups:

    - Full-time Employment
    - Part-time EMployment
    - Independent Contractor 
    - I prefer not to say / Others

```python
# Define the function
def clean_employment(x):
    if x in ["Employed, full-time",
            "Employed, full-time;Independent contractor, freelancer, or self-employed",
             "Employed, full-time;Employed, part-time",
             "Employed, full-time;Independent contractor, freelancer, or self-employed;Employed, part-time",
             "Employed, full-time;Independent contractor, freelancer, or self-employed;Retired",
             "Employed, full-time;Retired"
            ]:
        return "Full-time Employment"
    elif  x in ["Employed, part-time",
                "Employed, part-time;Retired"
                ]:
        return "Part-time Employment"
    elif  x in ["Independent contractor, freelancer, or self-employed",
                "Independent contractor, freelancer, or self-employed;Employed, part-time"
                ]:
        return "Independent Contractor"
    else:
        return "I prefer not to say / Others"

# Apply the function to the 'Employment' column in the DataFrame
df['Employment'] = df['Employment'].apply(clean_employment)
```

## C. Modeling

- **Label Encoding**

```python
le_Country = LabelEncoder()
df['Country'] = le_Country.fit_transform(df['Country'])
```

```python
le_Education = LabelEncoder()
df['Education'] = le_Education.fit_transform(df['Education'])
```

```python
le_Employment = LabelEncoder()
df['Employment'] = le_Employment.fit_transform(df['Employment'])
```

```python
le_Gender = LabelEncoder()
df['Gender'] = le_Gender.fit_transform(df['Gender'])
```

```python
df.head(5)
```

Our data is now displayed in readable format of ML model.

- **Split data**

```python
X = df.drop(["Salary", "Age", "Year_of_Code"], axis=1) # Only keey quality predictors for the model
y = df["Salary"]
```

- **Fit model**

```python
# Fit the model
xgb_model = xgb.XGBRegressor(objective="reg:squarederror", random_state=0)
xgb_model.fit(X, y.values)
```

```python
# Predict
y_pred = xgb_model.predict(X)
```

```python
# Measure the performance
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

## D. Pack the model

```python
data = {"model": xgb_model,
        "le_country": le_Country,
        "le_education": le_Education,
        "le_employment": le_Employment,
        "le_gender": le_Gender
       }
with open('Saved_model.pkl', 'wb') as file:
    pickle.dump(data, file)
```
